Un modulo es codigo escrito por otra persona, que nos sirve para incorporar a tu codigo. random... Python aislado para funcionar con determinado proyecto. 

Para activar o crear un entorno vitual: 
1. vamos a la consola
2. estamos en nuestra carpeta y escribimos py -m venv venv
3. Automaticamente se crea una carpeta que contiene (include, lib, scripts..)
4. Activacion en la consola: C:\Users\mor96198\pythonbasic>.\venv\Scripts\activate 
5. Para usar un python clonado y para dejarlo activado se crea un alias del comando anterior
6. Para desactivar el entorno vitual: deactivate
7. alias avenv=.\venv\Scriptd\activate    ----->Para que sea mas sencillo activar el entorno virual

# Listas y diccionarios anidados

Puedo tener listas que guarden diccionarios y diccionarios que guarden listas


```python
def run():
    squares = []
    for i in range(1,101):
        squares.append(i**2)
    
    print(squares)

if __name__ == '__main__':
    run()
```

```python
# def run():
#     thirds = []
#     for i in range(1,101):
#         if i % 3 != 0:
#             thirds.append(i**2)
#     print(thirds) Este problema se puede resolver todo con list comprenhensions:

    squares = [i**2 for i in range (1,101) if i % 3 != 0] #Se hizo lo mismo en una sola linea de codigo
    print(squares)

if __name__ == '__main__':
    run()
```

# List comprenhensions [element for elemnt in iterable if condition]

```python
def run():
    challenge = [i for i in range(1,99999) if i % 4 == 0 and i % 6 == 0 and i % 9 == 0]
    print (challenge)
    
if __name__ == '__main__':
    run()
```

# Dictionary comprehensions

```python
#Lo mismo que se hacen las listas se puede hacer en los diccionarios. 
def run():
#     my_dict = {}
    
#     for i in range(1,101):
#         if i % 3 != 0:
#             my_dict[i] = i**3
#Todo esto se puede hacer en una sola linea:

    my_dict = {i: i**3 for i in range(1,101) if i % 3 != 0}
        
    print (my_dict)
if __name__ == '__main__':
    run()
```

{key:value for value in iterable if condition}
1. key value: representa a cada una de las llaves y valores a poner en el nuevo diccionario. 
2. for value in iterable: Ciclo a partir del cual se extraeran elementos de cualquier iterable
3. if condition: Condicion final para filtrar los elementos del ciclo

```python
#Reto
def run():
    
    my_dict = {i: i**0.5 for i in range (1,1001)}
    
    print (my_dict)
if __name__ == '__main__':
    run()
```

# Funcion lambda

Funciones sin nombre, anonimas. No tienen un identificador. 
lambda argumentos:expresion (Linea de codigo que hace algo). 
Las funciones lambda tiene una sola linea de codigo. 

identificador: palindrome
argumento: lambda string
expresion: string == string[::-1]
Ver siguiente ejemplo:


```python
#Lambda function retorna un objeto de tipo funcion
palindrome = lambda string: string == string [::-1]
print(palindrome('ana'))

#Si fuera con funciones definidas se veria asi:
# def palindrome(string):
#     return string == string[::-1]
# print(palindrome('ana'))
```

# High order functions: filter, map and reduce

```python
#Filter
my_list = [1,4,5,6,9,13,19,21]
odd = list(filter(lambda x: x%2 != 0, my_list))
print (odd)
```

```python
#map
my_list = [1,2,3,4,5]
# squares = [i**2 for i in my_list]
# print(squares) Solution1
odd = list(map(lambda x: x**2, my_list))
print (odd) #Soluction 2

    
```

```python
#Function Reduvr
from functools import reduce

my_list = [2,2,2,2,2]
#First solution
# all_multiplied = 1
# for i in my_list:
#     all_multiplied = all_multiplied * i
# print(all_multiplied)
#Second Solution

all_multiplied = reduce(lambda a, b: a * b, my_list)
print(all_multiplied)
# odd = reduce(my_list)
# print(odd)
```

# Los errores en el codigo

1. KeyboardInterrupt = cuando oprimes ctrl + c 
2. Key error = inentamos acceder a un dicctionario con una llave que no existe
3. Indexerrror = cuando el elemento q intentamos acceder no esta en la lista
4. FileNotFoundError = Cuando no encuentra un archivo
5. ZeroDivisionError = cuando intentamos dividir algun numero sobre cero
6. Importerror = intentamos importar algo que falla en un modulo

Traceback (most recent call last):

Se lee desde el final hasta el principio 
 - Ultima linea la explicacion de la excepcion
 - penultima linea: el archivo, la linea y el modulo donde sucede el error
 - la primera linea: De donde parte el error
